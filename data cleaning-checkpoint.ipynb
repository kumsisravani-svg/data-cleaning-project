```python
import pandas as pd

df = pd.read_csv("amazon sale report.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 22'])
```

```python
df['Amount'] = df['Amount'].fillna(0)
```

```python
df['currency'] = df['currency'].fillna('Unknown')
df['Courier Status'] = df['Courier Status'].fillna('Unknown')
df['fulfilled-by'] = df['fulfilled-by'].fillna('Unknown')
```

```python
df = df.drop(columns=['promotion-ids'])
```

```python
df = df.dropna(subset=['ship-city', 'ship-state'])
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
```

```python
df['Date'].head()
df['Date'].dtype
```

```python
import matplotlib.pyplot as plt
df['Status'].value_counts().plot(kind='bar')
plt.title("Order Status Distribution")
plt.show()
```

```python
df.groupby('Category')['Amount'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
df['ship-state'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 States by Orders")
plt.show()
```

```python
df.groupby('Date')['Amount'].sum().plot()
plt.title("Sales Trend Over Time")
plt.show()
```

```python
df['B2B'].value_counts().plot(kind='bar')
plt.title("B2B vs B2C Orders")
plt.show()
```

```python
Key Insights from Data Analysis
The majority of orders fall under the “Shipped” status, indicating successful order processing and fulfillment.
Among all product categories, “Set” and “Kurta” contribute the most to total sales, showing strong customer preference for these categories.
The highest number of orders are from Maharashtra, followed by Karnataka, suggesting these regions have the highest demand.
Sales trend analysis shows a significant peak during the month of May, indicating a possible seasonal increase or promotional impact.
The dataset is dominated by non-B2B (False) transactions, while B2B orders are very minimal, showing that most customers are individual buyers rather than businesses.

    
```
